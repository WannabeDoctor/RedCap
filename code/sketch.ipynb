```python
from dataclasses import dataclass

@dataclass
class Patient:
    id: str

```

```python
@dataclass
class Lab:
    crp: float
    sodium: float
    potassium: float
    
    def PrintLab(self):
        print(f"CRP: {self.crp}, Na+: {self.sodium}, K+: {self.potassium}")
    
```

```python
@dataclass
class Patient:
    id: str
    lab: Lab
    bmi: float
    age: int
    gender: str
    
    def PrintLab(self):
        if self.lab:
            self.lab.PrintLab()
```
